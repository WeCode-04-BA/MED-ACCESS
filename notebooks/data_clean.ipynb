# Cleaning medical datasets
* Clean the diseases and symptoms dataset
* Look at the symptoms description, precaution, and severity datasets
* Make sure every dataset is cleaned and prepared

## Diseases and symptoms 

```python
import pandas as pd
diseases = pd.read_csv("./Data/dataset.csv")
```

```python
#view first 10 rows 
diseases[:10]
```

## The data is very messy 
* TO DO:
  1. Strip spaces and lowercase
  2. Remove duplicates
  3. Drop empty symptoms columns


```python
#Strip space and lowercase,replace spaces
for col in diseases.columns:
    diseases[col] = diseases[col].astype(str).str.strip().str.lower().str.replace(r"[\s_]+" , "_",regex =True)
    diseases[col] = diseases[col].replace("nan",pd.NA)
    
#view first 10 rows
diseases[:10]
```

## Now the data is more readable 
* remove duplicates

```python
diseases = diseases.drop_duplicates()
diseases[:10]
```

## So far so good, removed duplicates
* Now we will drop  empty symptom columns
* We will  try to drop columns with 80% missing values 

```python

threshold = len(diseases) * 0.2
diseases = diseases.dropna(axis=1, thresh= threshold)

#view first 15 rows
diseases[:15]
```

```python
#check missing values
diseases.isna().sum()
```

```python
# check duplicates
diseases.duplicated().sum()
```

## Done with diseases and symptoms dataset
* Export the dataset

```python
diseases.to_csv("cleaned_dataset.csv", index=False)
```
